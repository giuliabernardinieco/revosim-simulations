### Generating a 2D sinusoidal grating picture 

```python
import numpy as np
import matplotlib.pyplot as plt
import PIL 
```

```python


numbers = [50]
num_rows = 2
num_cols = len(numbers) // num_rows + 1

plt.figure(figsize=(15, 8))

for i, num in enumerate(numbers, start=1):
    x = np.arange(0, 100, 1)
    print(x)
    X, Y = np.meshgrid(x, x)
    
    # Create an array with constant red and green channels and a varying blue channel
    image = np.zeros((100, 100, 3))
    print(np.shape(image))
    image[:, :, 0] = 0  # Red channel
    image[:, :, 1] = 0  # Green channel
    for j in range(0,100):
        image[:, j, 2] = (np.sin(2.0 * np.pi * (j*1.0) / float(num)) + 1.0)/2.0  # Varying blue channel
        print(image[0, j, 2])
    
    plt.subplot(num_rows, num_cols, i)
    plt.axis('off')
    plt.imshow(image)
    plt.title(f'wavelength: {num}')
    break

plt.tight_layout()
plt.show()
```
