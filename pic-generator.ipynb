### Generating a 2D sinusoidal grating picture 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating 1D sine wave

x = np.arange(0, 200, 1)
# signal = amplitude * np.sin(2 * np.pi * x / wavelength)
y1 = np.sin(2 * np.pi * x / 100)
y2 = np.sin(2 * np.pi * x / 200)
y3 = np.sin(2 * np.pi * x / 50)
y4 = np.sin(2 * np.pi * x / 25)
y5 = np.sin(2 * np.pi * x / 10)
y = y1 + y2 + y3 + y4 + y5

plt.subplots(figsize=(12, 6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
#plt.plot(x, y, color = 'black', linewidth=2)
plt.show()
```

```python
numbers = [110]
num_rows = 2
num_cols = len(numbers) // num_rows + 1

plt.figure(figsize=(15, 8))

for i, num in enumerate(numbers, start=1):
    x = np.arange(0, 100, 1)
    X, Y = np.meshgrid(x, x)
    
    # Create an array with constant red and green channels and a varying blue channel
    image = np.zeros((100, 100, 3))
    image[:, :, 0] = 0  # Red channel
    image[:, :, 1] = 0.5  # Green channel
    image[:, :, 2] = (np.sin(2 * np.pi * X / num) + 1)/2  # Varying blue channel
    
    plt.subplot(num_rows, num_cols, i)
    plt.axis('off')
    plt.imshow(image)
    plt.title(f'wavelength: {num}')

plt.tight_layout()
plt.show()
```

```python
# Generate and save the pics

for num in numbers:
    x = np.arange(0, 100, 1)
    X, Y = np.meshgrid(x, x)
    image = np.zeros((100, 100, 3))
    image[:, :, 0] = 0  # Red channel
    image[:, :, 1] = 0.5  # Green channel
    image[:, :, 2] = (np.sin(2 * np.pi * X / num) + 1)/2  # Varying blue channel
    
    plt.figure(figsize=(1.3, 1.3))
    plt.axis('off')
    plt.imshow(image)
    plt.savefig(f'w{num}.png', dpi=100, bbox_inches='tight', pad_inches=0, transparent=True)
    plt.close()  
```
