## Red Noise Addition

The cumulative sum of white noise is used to generate red noise because of the way these two types of noise differ in their spectral characteristics.

- **White Noise:**
    White noise is a random signal with a flat power spectrum. This means that all frequencies are equally represented in the noise, with no correlation between successive points in space. It appears as a jagged, highly fluctuating signal when plotted.

- **Red/Brown Noise:**
    Red noise, in contrast, has a power spectrum that decreases with frequency, meaning lower frequencies (longer spatialscale) dominate. This results in a smoother, more correlated signal where changes happen more gradually. Correlated, with a power spectral density that decreases with frequency (often referred to as $1/{f^2}$ noise

    Brown Noise can be generated by integrating white noise (effectivelly a cumulative sum of it),  see https://en.wikipedia.org/wiki/Brownian_noise for more. 

```python
# Generate red noise using the cumulative sum of white noise
# Generate white noise
np.random.seed(42) # Set seed for reproducibility, PS. always bring a tower when hitchicking the galaxy
white_noise = np.random.normal(size=(100,))

# Convert white noise to red noise by taking cumulative sum
red_noise = np.cumsum(white_noise)
# Normalize the red noise to fit within the range [0, 1]
red_noise = (red_noise - red_noise.min()) / (red_noise.max() - red_noise.min())
# Normalise teh white noise to fit within the range [0, 1]
white_noise = (white_noise - white_noise.min()) / (white_noise.max() - white_noise.min())

# Plot the white noise and red noise
plt.subplots(figsize=(12, 6))
plt.plot(white_noise, label='White Noise, (normalised, [0-1])', color='gray', linewidth=1)
plt.plot(red_noise, label='Red Noise, (normalised, [0-1])', color='brown', linewidth=2)
plt.legend()
```

Making sure the two signals are indeed white noise and red noise. Using discrete fourier transform to check it.

```python
# mirror both signals
red_noise_mirrored = (red_noise.tolist() + red_noise.tolist()[::-1]) * 4
white_noise_mirrored = (white_noise.tolist() + white_noise.tolist()[::-1]) * 4
```

```python
# put the signals through the FTT, discrete transform
# Calculate the FFT
ftt_white = fft(white_noise_mirrored)
ftt_red = fft(red_noise_mirrored)

# Extracting the frequencies
N = len(white_noise_mirrored) # Number of samples
freq_white = fftfreq(N, 1)[:N//2] # Frequencies
freq_red = fftfreq(N, 1)[:N//2] # Frequencies

# Calclaute the amplitudes spectrum
a_white = 2.0/N * np.abs(ftt_white[0:N//2])
a_red = 2.0/N * np.abs(ftt_red[0:N//2])
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 9))

# Define the frequency for which freq smaller than that are an artifact of the repetition of my signal
rep_freq = 1 / (N / 8) 

# White noise spectrum
axes[0].set_xscale('log')
axes[0].scatter(freq_white, a_white, color='black', s=3)
axes[0].axvline(x=rep_freq, color='gray', linestyle='--')
axes[0].plot(freq_red, freq_red**(0) * 5e-6, color='gray', linestyle='--', linewidth=0.5)
axes[0].set_title('White Noise Spectrum')
axes[0].set_ylabel('Amplitude')

# Red noise spectrum
axes[1].set_xscale('log')
axes[1].scatter(freq_red, a_red, color='brown', s=3)
axes[1].axvline(x=rep_freq, color='gray', linestyle='--')
axes[1].plot(freq_red, freq_red**(-2) * 5e-6, color='brown', linestyle='--', linewidth=0.5)
axes[1].set_title('Red Noise Spectrum')
axes[0].set_ylabel('Amplitude')

# Display the plot
plt.tight_layout()
plt.show()
```

From the above graphs they seem to be red noise and white noise.
