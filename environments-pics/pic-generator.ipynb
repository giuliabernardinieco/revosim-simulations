## Red Noise Addition

The cumulative sum of white noise is used to generate red noise because of the way these two types of noise differ in their spectral characteristics.

- **White Noise:**
    White noise is a random signal with a flat power spectrum. This means that all frequencies are equally represented in the noise, with no correlation between successive points in space. It appears as a jagged, highly fluctuating signal when plotted.

- **Red/Brown Noise:**
    Red noise, in contrast, has a power spectrum that decreases with frequency, meaning lower frequencies (longer spatialscale) dominate. This results in a smoother, more correlated signal where changes happen more gradually. Correlated, with a power spectral density that decreases with frequency (often referred to as $1/{f^2}$ noise

    Brown Noise can be generated by integrating white noise (effectivelly a cumulative sum of it),  see https://en.wikipedia.org/wiki/Brownian_noise for more. 

```python
# Generate red noise using the cumulative sum of white noise
# Generate white noise
np.random.seed(42) # Set seed for reproducibility, PS. always bring a tower when hitchicking the galaxy
white_noise = np.random.normal(size=(100,))

# Convert white noise to red noise by taking cumulative sum
red_noise = np.cumsum(white_noise)
# Normalize the red noise to fit within the range [0, 1]
red_noise = (red_noise - red_noise.min()) / (red_noise.max() - red_noise.min())
# Normalise teh white noise to fit within the range [0, 1]
white_noise = (white_noise - white_noise.min()) / (white_noise.max() - white_noise.min())

# Plot the white noise and red noise
plt.subplots(figsize=(12, 6))
plt.plot(white_noise, label='White Noise, (normalised, [0-1])', color='gray', linewidth=1)
plt.plot(red_noise, label='Red Noise, (normalised, [0-1])', color='brown', linewidth=2)
plt.legend()
```

Making sure the two signals are indeed white noise and red noise. Using discrete fourier transform to check it.

```python
# mirror both signals
red_noise_mirrored = (red_noise.tolist() + red_noise.tolist()[::-1]) * 4
white_noise_mirrored = (white_noise.tolist() + white_noise.tolist()[::-1]) * 4
```

```python
# put the signals through the FTT, discrete transform
# Calculate the FFT
ftt_white = fft(white_noise_mirrored)
ftt_red = fft(red_noise_mirrored)

# Extracting the frequencies
N = len(white_noise_mirrored) # Number of samples
freq_white = fftfreq(N, 1)[:N//2] # Frequencies
freq_red = fftfreq(N, 1)[:N//2] # Frequencies

# Calclaute the amplitudes spectrum
a_white = 2.0/N * np.abs(ftt_white[0:N//2])
a_red = 2.0/N * np.abs(ftt_red[0:N//2])
```

```python
# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 9))

# Define the frequency for which freq smaller than that are an artifact of the repetition of my signal
rep_freq = 1 / (N / 8) 

# White noise spectrum
axes[0].set_xscale('log')
axes[0].scatter(freq_white, a_white, color='black', s=3)
axes[0].axvline(x=rep_freq, color='gray', linestyle='--')
axes[0].plot(freq_red, freq_red**(0) * 5e-6, color='gray', linestyle='--', linewidth=0.5)
axes[0].set_title('White Noise Spectrum')
axes[0].set_ylabel('Amplitude')

# Red noise spectrum
axes[1].set_xscale('log')
axes[1].scatter(freq_red, a_red, color='brown', s=3)
axes[1].axvline(x=rep_freq, color='gray', linestyle='--')
axes[1].plot(freq_red, freq_red**(-2) * 5e-6, color='brown', linestyle='--', linewidth=0.5)
axes[1].set_title('Red Noise Spectrum')
axes[1].set_ylabel('Amplitude')

# Display the plot
plt.tight_layout()
plt.show()
```

## Perlin Noise addition

Caracteristic of Perlin Noise

- **Gradient Noise**:
    Unlike pure random noise, which appears as a chaotic, "static" pattern, Perlin noise is smooth and continuous. This is achieved by calculating gradients at grid points and interpolating between them. The result is a pattern that looks more natural and organic.

- **Smoothness**:
    Perlin noise produces smooth transitions between values, creating a flow that avoids sharp edges or abrupt changes. This is ideal for generating textures that mimic natural phenomena, like clouds, wood grain, or terrain.

- **Multidimensional**:
    Perlin noise can be generated in 1D, 2D, 3D, or higher dimensions, making it versatile for different applications. For example, 2D Perlin noise can create textures, while 3D Perlin noise can be used for volumetric effects like clouds.
    
- **Tiling and Repetition**:
    Perlin noise can be made to tile seamlessly, which is useful for textures that need to repeat without visible seams.

#### How does perlin noise work from the perlin noise library

The following parameters control the characteristics of the Perlin noise generated by the `noise.pnoise2` function. They determine how the noise appears, affecting the texture, detail, and variation of the pattern.

##### 1. `scale`
Controls the "zoom" level of the noise, effectively determining how stretched or compressed the noise pattern is across the image.
- **Low Scale Value** (e.g., 2.0): The noise pattern appears zoomed in, resulting in large, smooth regions of similar values with less detail.
- **High Scale Value** (e.g., 50.0): The noise pattern appears zoomed out, with more frequent and smaller changes across the image, resulting in a finer, more detailed texture.

##### 2.`octaves`
Determines the number of layers of noise (or frequencies) combined to produce the final noise pattern. Each octave is a noise function at a different frequency, stacked on top of each other.
- **Fewer Octaves** (e.g., 1 or 2): Results in a simpler, smoother noise pattern with large features.
- **More Octaves** (e.g., 5 or 6): Adds complexity by layering additional noise at higher frequencies, producing a more detailed and natural-looking pattern.

##### 3. `persistence`
Controls the amplitude of each successive octave relative to the previous one. It determines how much influence each layer (octave) has on the final noise pattern.
- **Low Persistence** (e.g., 0.3): Each successive octave contributes less to the overall noise, leading to a smoother pattern dominated by the lower-frequency layers.
- **High Persistence** (e.g., 0.7): Higher-frequency octaves contribute more, adding finer details and more variation to the noise pattern.

##### 4. `lacunarity`
Controls the frequency of each successive octave relative to the previous one. Specifically, it multiplies the frequency by a factor with each octave.
- **Low Lacunarity** (e.g., 1.5): Results in each octave having a frequency only slightly higher than the previous one, leading to smoother transitions.
- **High Lacunarity** (e.g., 3.0): Results in a more significant increase in frequency between octaves, which adds more fine detail and sharper transitions between noise features.

##### Example of Their Combined Effect:
- **Low Scale, High Octaves, High Persistence, Low Lacunarity**: Produces a highly detailed pattern with gradual transitions between features.
- **High Scale, Low Octaves, Low Persistence, High Lacunarity**: Results in a simpler pattern with less detail and sharp transitions.

## Sintetic noise
I will now try to create red noise by combining multiple sin functions to create a more complex function.



```python

N = 128 #number of samples
dt = 1    #timestep
t = np.arange(0,N,dt)  #times

f1 = 1/128  #frequency
a1 = 1    #amplitude

f2 = 1/64 
a2 = 0.5

f3 = 1/32
a3 = 0.25

f4 = 1/16
a4 = 0.125

f5 = 1/8
a5 = 0.0625


z1 = a1 * np.sin(t*f1*np.pi*2.)
z2 = a2 * np.sin(t*f2*np.pi*2.) 
z3 = a3 * np.sin(t*f3*np.pi*2.)
z4 = a4 * np.sin(t*f4*np.pi*2.)
z5 = a5 * np.sin(t*f5*np.pi*2.)

z = z1 + z2 + z3 + z4 +z5

plt.plot(t,z1, color='orange', linewidth=0.5)
plt.plot(t,z2, color='pink', linewidth=0.5)
plt.plot(t,z3, color='green', linewidth=0.5)
plt.plot(t,z4, color='blue', linewidth=0.5)
plt.plot(t,z5, color='red', linewidth=0.5)
plt.plot(t,z, color='black', linewidth=2)
plt.xlim(0,128)
plt.show()
```

```python
zf = fft(z)  #fourier transform of z
zfreq = fftfreq(N, dt)[:N//2] #define frequencies for plotting. 
#this function above extract the Discrete Fourier Transform sample frequencies,
#N is our sample that we defined before, and d1 is our scalar: we want all the frequency from 0 up to N/2
##GB why //2?

plt.plot(zfreq, 2.0/N * np.abs(zf[0:N//2]))  #plot series in frequency domain (i.e. as function of frequency)

#plot frequencies and amplitudes used to generate original series

plt.plot(f1, a1, color='orange', marker='o')
plt.plot(f2,a2, color='pink', marker='o')
plt.plot(f3,a3, color='green', marker='o')
plt.plot(f4,a4, color='blue', marker='o')
plt.plot(f5,a5, color='red', marker='o')

plt.ylabel('Amplitude')
plt.xlabel('Frequency')
```
